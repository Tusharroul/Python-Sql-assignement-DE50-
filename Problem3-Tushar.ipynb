```python
print("The average profit of the company in a month is", sum(df.total_profit)/len(df))
```

# 6. Get the most sold item of the company 

```python
n=[]
for i in l:
    n.append(max(df[i]))
for i in l:
    if max(df[i])==max(n):
        print("The most sold item is", i)
```

# 7. Read Total profit of all months and show it using a line plot 

```python
plt.plot(df.month_number,df.total_profit)
plt.title('total profit of all the months')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

# 8. Read all product sales data and show it using a Multiline Plot

```python
plt.plot(df.facecream, df.total_profit, label = "facecream", linestyle="-")
plt.plot(df.facewash, df.total_profit, label = "facewash", linestyle="--")
plt.plot(df.toothpaste, df.total_profit, label = "toothpaste", linestyle="-")
plt.plot(df.bathingsoap, df.total_profit, label = "bathingsoap", linestyle="--")
plt.plot(df.shampoo, df.total_profit, label = "shampoo", linestyle="-")
plt.plot(df.moisturizer, df.total_profit, label = "moisturizer", linestyle="--")

plt.legend()
plt.show()
```

# 9. Read toothpaste sales data of each month and show it using a scatter plot

# 10. Read face cream and face wash product sales data and show it using the bar chart

```python
plt.bar(df.month_number, df.facecream)
plt.title('face cream sold in all the months')
plt.xlabel('month')
plt.ylabel('facecream')
plt.show()
```

```python
plt.bar(df.month_number, df.facewash)
plt.title('face wash sold in all the months')
plt.xlabel('month')
plt.ylabel('facewash')
plt.show()
```

# 11. Read Bathing soap and face wash data of all months and display it using a Subplot
